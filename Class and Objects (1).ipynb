# Class
* A class is a user-defined blueprint or prototype from which objects are created.
* Classes provide a means of bundling data and functionality together. 
* Creating a new class creates a new type of object, allowing new instances of that type to be made.
* Each class instance can have attributes attached to it for maintaining its state. 
* Class instances can also have methods (defined by their class) for modifying their state.
* Syntax: Class Definition
   * class ClassName:
   *  #Statement
* The class creates a user-defined data structure, which holds its own data members and member functions, which can be accessed and used by creating an instance of that class.
* Some points on Python class:
  * Classes are created by keyword class.
  * Attributes are the variables that belong to a class.
  * Attributes are always public and can be accessed using the dot (.) operator. Eg.: My class.Myattribute

## Instant Attribute
* Instance attributes are defined within the methods of the class and accessed using the self keyword. 
* Instance attributes represent the specific characteristics or state of each object1. 
* The most common method used to define instance attributes is the init () function, which is the constructor of the class2.
![](https://miro.medium.com/v2/resize:fit:1400/1*mgYpKIBJGf1uJu7ViyM9Fg.png)

```python
class Student: 
    def __init__(self, name, course): 
        self.name = name # instance attribute
        self.course = course # instance attribute
```

```python
x = Student('ayesha','python') 
```

```python
x.name
```

```python
x.course
```

* name and course are instance attributes that are defined in the init () function and assigned to each object using the self parameter. 
* Each object has its own copy of the instance attributes and can have different values for them.

```python
#instance attribute
class stu:
    def __init__(self,passmarks,*marks):
        self.passmark=passmarks
        self.res='Fail' if min(marks)<self.passmark else 'pass'
```

```python
x=stu(40,*[20,30,40,50])
```

### Static Attribute 
* A static attribute in Python is a variable or a field that belongs to the class itself, not to any instance of the class. 
* It is shared by all instances of the class and can be accessed without creating an object of the class. 
* Static attributes are useful for storing global or constant values that are relevant for the class.

```python
class stu:
    passmark=40
    def __init__(self,*marks):
        self.marks=marks
        self.res='Fail' if min(marks)<self.passmark else 'pass'
```

```python
x=stu(*[20,30])
```

```python
x.res
```
