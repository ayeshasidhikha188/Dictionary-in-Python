# Dictionary
* Dictionaries are key & value pairs enclosed in { }
* key & value are seperated by ':'  while key and value pairs are sperated by ','
* keys are mapped to their values 
* In dictionaries , keys are unique, we cant have duplicate keys, we can duplicate values
* Dictionaries are unordered & mutable
* { } is an empty dictionary
* dict is class of dictionary 
* Dictionary holds key:value pair. 


```python
#dictionary, key;value
x={1:[1,2,3],'ayesha':(4,5,6)}  #'1' is int and 'ayesha'is string,('f','g') is tuple
print(x,type(x))
```

### Creating a Dictionary
* In Python, a dictionary can be created by placing a sequence of elements within curly {} braces, separated by ‘comma’. * Dictionary holds pairs of values, one being the Key and the other corresponding pair element being its Key:value.
* Values in a dictionary can be of any data type and can be duplicated, whereas keys can’t be repeated and must be immutable


```python
x={1:[1,2,3],'ayesha':(4,5,6)} 
print(x,type(x))
```

## creating an empty  dictionary
* Empty dictionary can also be created by using dict(). 
* An empty dictionary can be created by just placing to curly braces{}. 

```python
# using {} :
x={}
print(x,type(x))
```

```python
#using class dict()
x=dict()
print(x,type(x))
```

### Range 
* range()populates a numbers within a given range 
* syntax for range() is : range(start,end,step)
  * start is starting number 
  * end is till end , not including end 
  * step is step size 
* default values of start is zero and step is +1

```python
r= range(1,11)  # till end means upto 10
print(r,type(r),hex(id(r)),sep='\n')
```

### Methods in Dictionary
* Dictionary is like a map that is used to store data in the form of a key: value pair.
* Python provides various built-in functions to deal with dictionaries
1. clear()

* Removes all items from the dictionary

2. copy()

* Returns a shallow copy of the dictionary

3. fromkeys()

* Creates a dictionary from the given sequence

4. get()

* Returns the value for the given key

5. items()

* Return the list with all dictionary keys with values

6. keys()

* Returns a view object that displays a list of all the keys in the dictionary in order of insertion

7. pop()

* Returns and removes the element with the given key

8. popitem()

* Returns and removes the key-value pair from the dictionary

9. setdefault()

* Returns the value of a key if the key is in the dictionary else inserts the keywith a value to the dictionary

10. values()

* Updates the dictionary with the elements from another dictionary

11. update()

* Returns a list of all the values available in a given dictionary
